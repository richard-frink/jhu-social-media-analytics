```python
# creating a mapping of users we need to query the api for - going to see if they are connected
userFriendshipMapping = []

for user in users:
    for user2 in users:
        if user != user2:
            if user < user2:
                if [user, user2] not in userFriendshipMapping:
                    userFriendshipMapping.append([user, user2])
            else:
                if [user2, user] not in userFriendshipMapping:
                    userFriendshipMapping.append([user2, user])

for mapping in userFriendshipMapping:
    print(mapping)

print(len(userFriendshipMapping))
```

```python
friendshipResults = []

for user1, user2 in userFriendshipMapping:
    relationship = api.ShowFriendship(source_user_id=user1, target_user_id=user2)
    friendshipResults.append(relationship)
    print(relationship)
    print("retrieved friendship information for -> " + str(user1) + "  --  " + str(user2))
    print("waiting for 65 seconds to avoid 15 calls per 15 minutes rate limit...")
    time.sleep(65)
    
# checks the friendship between two users
#print(api.ShowFriendship(source_user_id=27260086, target_user_id=23375688))
```

```python
# building the relationships with our data

# the relationship object has
# id/screen name,
# following - meaning following the other user, and
# followed by - meaning followed by the other user

# ready our edge list
edgeList = []

# add initial nodes
[edgeList.append([userScreenNamesToIndex[u.screen_name], userScreenNamesToIndex[u.screen_name], 0]) for u in fullUsers]

for dict in friendshipResults:
    friendship = dict["relationship"]
    sourceScreenName = friendship["source"]["screen_name"]
    targetScreenName = friendship["target"]["screen_name"]
    sourceFollowingTarget = 0
    targetFollowingSource = 0
    resultString = "source id: " + friendship["source"]["id_str"] + " screenName: " + friendship["source"]["screen_name"]
    if friendship["source"]["following"]:
        resultString += " is following "
        sourceFollowingTarget = 1
    else:
        resultString += " is not following "
    resultString += " target id: " + friendship["target"]["id_str"] + " screenName: " + friendship["target"]["screen_name"]
    print(resultString)
    resultString = "target id: " + friendship["target"]["id_str"] + " screenName: " + friendship["target"]["screen_name"]
    if friendship["target"]["following"]:
        resultString += " is following "
        targetFollowingSource = 1
    else:
        resultString += " is not following "
    resultString += " source id: " + friendship["source"]["id_str"] + " screenName: " + friendship["source"]["screen_name"]
    print(resultString)
    print()
    edgeList.append([userScreenNamesToIndex[sourceScreenName], userScreenNamesToIndex[targetScreenName], sourceFollowingTarget])
    edgeList.append([userScreenNamesToIndex[targetScreenName], userScreenNamesToIndex[sourceScreenName], targetFollowingSource])
```

```python
for edge in edgeList:
    print(edge)
```

```python
import numpy as np

n = max(max(user1 + 1, user2 + 1) for user1, user2, following in edgeList) # Get size of matrix
matrix = np.zeros((n, n))
for user1, user2, following in edgeList:
    matrix[user1][user2] = following # Convert to 0-based index.

# the meaning is the row index vs the column index --- row-user following the column-user / true or false
for row in matrix:
    print(row)
```

```python
rows = []

firstRow = ""
for user in userIntToScreenNames:
    firstRow += "," + user

rows.append(firstRow)

i = 0
for row in matrix:
    rowString = userIntToScreenNames[i]
    for element in row:
        rowString += "," + str(int(element))
    rows.append(rowString)
    i += 1

# rows are built for csv, time to write them
f = open("twitter_relationships.csv", "+a")
for row in rows:
    f.write(row + "\n")

f.close()
```
